```python
from binarytree import Node as Visual

class Node:
    
    DIRECTIONS = {-1: 'left', 1: 'right'}
    
    def __init__(self, data: int, parent=None, moves=[]):
        self.left     = None
        self.right    = None
        self.parent   = parent
        self.data     = data
        self.leftCnt  = 0
        self.rightCnt = 0
        self.content  = {
            'data': lambda: self.data, 
            'balance': lambda: self.rightCnt - self.leftCnt
        }
        self.balance(moves)
        
    def isRoot(self) -> bool:
        return not self.parent
        
    def isUnbalanced(self) -> bool:
        return abs(self.content['balance']()) > 1
    
    def detachParent(self) -> Node:
        parent = self.parent
        if self == self.parent.left:
            self.parent.left = None
        else:
            self.parent.right = None
        self.parent = None
        
        return parent
        
    def balance(self, moves: list):
        if len(moves) >= 2:
            last2Moves = [moves[-1]] + [moves[-2]]
            if sum(last2Moves) == 0:
                pass
            else:
                firstParent  = self.parent
                if firstParent.parent.isUnbalanced():
                    if firstParent.parent.isRoot():
                        detachedSecondParent = firstParent.detachParent()
                        firstParent.right, detachedSecondParent.parent = detachedSecondParent, firstParent
                    else:
                        detachedThirdParent  = secondParent.detachParent()
                        detachedSecondParent = firstParent.detachParent()
                    #print(secondParent.content['balance']())
        
    def insert(self, data: int, moves=None, root=None) -> tuple:
        if root is None:  root  = self
        if moves is None: moves = []
            
        for sign in self.DIRECTIONS:
            child = self.DIRECTIONS[sign]
            if sign * data >= sign * self.content['data']():
                setattr(self, child+'Cnt', getattr(self, child+'Cnt') + 1)
                moves.append((sign, )
                if getattr(self, child) is None:
                    setattr(self, child, Node(data, self, moves))
                    if self.isRoot():
                        print([self.content['data']()])
                        root = self
                else:
                    branch = getattr(self, child)
                    root = branch.insert(data, moves, root)
                break
                
        return root
            
    def printTree(self, contentKey: str, visual=None) -> Visual:
        if visual is None: 
            visual = Visual(self.content[contentKey]())
            
        for sign in self.DIRECTIONS:
            child = self.DIRECTIONS[sign]
            if getattr(self, child):
                branch = getattr(self, child)
                setattr(visual, child, Visual(branch.content[contentKey]()))
                branch.printTree(contentKey, getattr(visual, child))
                
        return visual
```
